# Decision Tree (Classification)

## Introduction
* Decision trees can be used for classification or regression


## Entropy
* In classification problems, decision trees try to minimize **entropy** at each leaf node
    * The "branched-out" leaf nodes will (ideally) contain examples from one class
    * Entropy = 0 if the leaf node contains examples from same class
    * Entropy = 1 if examples in the node are evenly split between classes

## Information Gain
* Information gain = (entropy of parent) - (weighted average of entropy of children)
* Decision trees try to maximize information gain when deciding where to split the leaves

## Parameters
* min_samples_split: minimum size of the leaf node
    * Smaller value produces more intricate boundary lines

## Considerations
### Use Cases

* Cancer classification

### Pros

### Cons

## Example: Iris Data
This example uses the wine dataset (from scikit-learn) to try and predict the wine type.

### Loading Data

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split

wine = load_wine()
x = wine.data
y = wine.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42) #saving 10% of the dataset for testing
```

### Creating a Decision Tree

```python
import matplotlib.pyplot as plt
from sklearn import tree

clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train, y_train) #fit the classifier to training data
score = clf.score(x_test, y_test)
print('The decision tree has {:.2%} accuracy.'.format(score))
plt.figure(figsize=[8,6])
tree.plot_tree(clf, max_depth=4)
plt.show()
```

### Optimizing the Decision Tree

```python
num_features = 14
used_features = num_features
for i in range(num_features-1,1,-1):
    _clf = tree.DecisionTreeClassifier()
    _x = x_train[:, :i]
    _clf.fit(_x, y_train)
    _score = clf.score(x_test, y_test)
    if _score > score:
        used_features = i
        score = _score
print('The decision tree is optimized by using the first {} features with an accuracy of {:.2%}.'.format(used_features, score))
```
