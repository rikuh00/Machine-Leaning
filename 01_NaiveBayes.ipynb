# Naive Bayes

## Introduction
* Naive Bayes is a **supervised classification** technique based on Bayes' Theorem
* The algorithm creates a **decision surface** to classify data based on **features** (characteristics)

## Bayes' Rule
* Bayes' rule calculates the probability of an outcome given a **prior**
    * e.g. probability of you selling an icecream (outcome) given the sunny weather (prior)
* Given multiple priors, the Naive Bayes algorithm determines which outcome is most likely
    


## Considerations
### Use Cases
* Text classification
    * e.g. figuring out the true identify of Robert Galbraith

### Pros
* Easy to implement
* Simple algorithm
* Can be applied to a large dataset

### Cons
* Does not consider word orders and phrases
    * e.g. "Chicago Bulls" is not recognized as a sports team and instead recognized as a city and an animal



## Example: Iris Data
This example uses the iris data set (from scikit-learn) to try and classify the type of iris based on petal length and width.

### Loading Data

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
x = iris.data[:, 2:] #only dealing with 2 features
y = iris.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=84) #saving 30% of the dataset for testing
```

### Creating the Classifier

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
_ = clf.fit(x_train, y_train) #fit the classifier to training data
```

### Plotting Data

```python
import numpy as np
import matplotlib.pyplot as plt
IRIS_LIST = ['Iris Setosa','Iris Versicolour','Iris Virginica']
COLOR_LIST = np.array(['tab:blue','tab:orange','tab:green'])
fig, ax = plt.subplots(figsize=[8,4])
ax.set_axisbelow(True)
ax.grid()

for i in range(3):
    low_lim = i*50
    high_lim = (i+1)*50
    ax.scatter(x[low_lim:high_lim, 0], x[low_lim:high_lim, 1], c=COLOR_LIST[i], label=IRIS_LIST[i])

ax.legend()
ax.set_xlabel('Petal Length [cm]')
ax.set_ylabel('Petal Width [cm]')
ax.set_title('Classifying Iris')
plt.show()
```

### Making Predictions

```python
l = 5 #length of petal [cm]
w = 1.5 #width of petal [cm]
outcome = clf.predict([[w, l]])[0]
print('Petal length: {}cm'.format(l))
print('Petal width: {}cm'.format(w))
print('This is an {}'.format(IRIS_LIST[outcome]))
```

### Evaluating the Classifier

```python
score = clf.score(x_test, y_test)
print('The classifier has {:.2%} accuracy.'.format(score))
```
